```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
heart_df = pd.read_csv('heartDieasePreprocess.csv')
heart_df.head()
```

```python
#Number of patients
fig,axes = plt.subplots(1,2,figsize = (10,5))
ax1 = heart_df.Disease.value_counts().plot(kind = 'bar',ax = axes[0])
ax1.set_title('Distribution of patients')
ax1.set_xticklabels(['Sick','Not Sick'])
ax2 = heart_df.Disease.value_counts().plot(kind = 'pie',autopct = '%.2f%%',labels = ['Sick','Not sick'],ax = axes[1])
ax2.set_title('Distribution of patients')
```

```python
# Seperate sick and non sick
patient = heart_df[heart_df['Disease'] == 1]
normal = heart_df[heart_df['Disease'] == -1]
```

### Gender and disease

```python
ax1 = plt.subplot(121)
sns.countplot(x = 'Sex',data = patient,ax = ax1)
ax1.set_xticklabels(['female','male'])
ax1.set_title('Distribution of Gender of patients')

ax2 = plt.subplot(122)
patient.Sex.value_counts().plot(kind = 'pie',autopct = '%.2f%%',labels = ['male','female'],ax = ax2)
ax2.set_title('Distribution of Gender of patients')
```

### Age and Disease

```python
# Distribution of age of patients
fig,axes = plt.subplots(2,1,figsize = (10,10))
ax1 = sns.countplot(x = 'Age',data = patient,ax = axes[0])

# 0-45Young   45-59Middle   60-100 Old
# Devide age
age_type = pd.cut(heart_df.Age,bins = [0,39,59,100],include_lowest = True,right = False,labels = ['Young','Middle','Old'])
# print(age_type)
age_target_df = pd.concat([age_type,heart_df['Disease']],axis = 1)
# print(age_target_df)
sns.countplot(x = 'Age',hue = 'Disease',data = age_target_df,ax = axes[1])
```

## Rest characteristic

```python
# resting_BP - Resting blood presure (in mm Hg on admission to the hospital)¶

fig,axes = plt.subplots(2,1,figsize = (10,10))
sns.set(color_codes = True)
resting_BP_normal = sns.histplot(normal['resting_BP'],kde = True,ax = axes[0])
axes[0].set_title("rBP for the normal")
resting_BP_normal.set_ylabel("Number")

resting_BP_patient = sns.histplot(patient['resting_BP'],kde = True,ax = axes[1])
axes[1].set_title("rBP for patients")
resting_BP_patient.set_ylabel("Number")
fig.tight_layout()
```

```python
# Serum_cholest	 - Cholesterol in mg/dl
fig,axes = plt.subplots(2,1,figsize = (10,10))
sns.set(color_codes = True)
plt.rcParams['font.sans-serif'] = ['SimHei']
Serum_cholest_normal = sns.histplot(normal['Serum_cholest'],kde = True,ax = axes[0])
axes[0].set_title("Normal Cholesterol Level")
Serum_cholest_normal.set_ylabel("Number")

Serum_cholest_patient = sns.histplot(patient['Serum_cholest'],kde = True,ax = axes[1])
axes[1].set_title("Patient's Cholesterol Level")
Serum_cholest_patient.set_ylabel("Number")
fig.tight_layout()
```

```python
# blood_sugar_exc120 - (Fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
# plt.figure(figsize = (7,7))
fig,axes = plt.subplots(1,2,figsize = (13,13))
blood_sugar_exc120_patient = patient['blood_sugar_exc120'].value_counts().plot(kind = 'pie',autopct = '%.2f%%',
                                       labels = ['FBS>120 mg/dl','FBS<=120 mg/dl'],ax = axes[0])
blood_sugar_exc120_patient.set_title('Patients\' FBS')

blood_sugar_exc120_normal = normal['blood_sugar_exc120'].value_counts().plot(kind = 'pie',autopct = '%.2f%%',
                                       labels = ['FBS>120 mg/dl','FBS<=120 mg/dl'],ax = axes[1])
blood_sugar_exc120_normal.set_title('Normal FBS')
plt.tight_layout()
```

```python
# Max_heart_rate - Maximum heart rate
fig,axes = plt.subplots(1,2,figsize = (13,8))
sns.set(color_codes = True)
plt.rcParams['font.sans-serif'] = ['SimHei']

Max_heart_rate_patient = sns.histplot(patient['Max_heart_rate'],kde = True,ax = axes[0])
Max_heart_rate_patient.set_title('Patients\' MHR' )


Max_heart_rate_normal = sns.histplot(normal['Max_heart_rate'],kde = True,ax = axes[1])
Max_heart_rate_normal.set_title('Normal MHR')
```

```python
# ind_exerc_angina - Exercise-induced Angina (1 = yes; 0 = no)
# plt.figure(figsize = (7,7))
fig,axes = plt.subplots(1,2,figsize = (13,13))

ind_exerc_angina_patient = patient['ind_exerc_angina'].value_counts().plot(kind = 'pie',autopct = '%.2f%%',
                                       labels = ['yes','no'],ax = axes[0])
ind_exerc_angina_patient.set_title('EIA for Normal ppl')

ind_exerc_angina_normal = normal['ind_exerc_angina'].value_counts().plot(kind = 'pie',autopct = '%.2f%%',
                                       labels = ['yes','no'],ax = axes[1])
ind_exerc_angina_normal.set_title('EIA for Patients')
```

```python
#Angina_type - chest pain type（1：typical，2：atypical，
#3：non-anginal，4：asymptomatic）
plt.figure(figsize = (7,7))

Angina_type = patient['Angina_type'].value_counts().plot(kind = 'pie',autopct = '%.2f%%',
                                       labels = ['Typical','Atypical','Non-anginal','Asymptomatic'])
Angina_type.set_title('Chest pain types for patients')
```

```python
#ST depression induced by exercise (cts) 
fig,axes = plt.subplots(1,2,figsize = (13,8))
sns.set(color_codes = True)

ST_dep_by_exerc_patient = sns.histplot(patient['ST_dep_by_exerc'],kde = True,ax = axes[0])
ST_dep_by_exerc_patient.set_title("ST depression induced by exercise")

ST_dep_by_exerc_normal = sns.histplot(normal['ST_dep_by_exerc'],kde = True,ax = axes[1])
ST_dep_by_exerc_normal.set_title("ST depression induced by exercise")
```

```python
# ST-segment peak slop 1ope
plt.figure(figsize = (7,7))
plt.rcParams['font.sans-serif'] = ['SimHei']
ind_for_slope_exerc = normal['ind_for_slope_exerc'].value_counts().plot(kind = 'pie',autopct = '%.2f%%',
                                       labels = ['flat','up','down'])
ind_for_slope_exerc.set_title('ST-segment peak slop 1ope')
```

```python
# num_vessels_fluro
# plt.figure(figsize = (7,7))
fig,axes = plt.subplots(1,2,figsize = (13,13))

num_vessels_fluro_patient = patient['num_vessels_fluro'].value_counts().plot(kind = 'pie',autopct = '%.2f%%',ax = axes[0])
num_vessels_fluro_patient.set_title('Number of vessels for patients')

num_vessels_fluro_normal = normal['num_vessels_fluro'].value_counts().plot(kind = 'pie',autopct = '%.2f%%',ax = axes[1])
num_vessels_fluro_normal.set_title('Normal number of vessels for patients')
```

```python
#Thal_defect 1 reversable fixed 0正常

plt.figure(figsize = (7,7))
plt.rcParams['font.sans-serif'] = ['SimHei']
Thal_defect = patient['Thal_defect'].value_counts().plot(kind = 'pie',autopct = '%.2f%%',
                                                        labels = ["reversable","fixed","normal"])
Thal_defect.set_title('Thalassemia')
```

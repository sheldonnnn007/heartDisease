```python
import pandas as pd
```

```python
heart = pd.read_csv('heartDisease.csv')
heart.head()
```

```python
# Sick?：0 NO，1 YES
disease = heart['Disease']
for x in disease.index:
    if x == -1:
        disease[x] = 0
heart_new = heart.drop('Disease', axis=1)
heart_new['Disease'] = disease
```

```python
# Angina type  1typical Angina 2atypical Angina 3non Angina 4asymptomatic
angina = heart_new['ind_typ_angina']
angina = angina.copy()
for x in heart_new.index:
    if heart_new.loc[x, 'ind_typ_angina'] == 1:  # typical Angina
        angina[x] = 1
    elif heart_new.loc[x, 'ind_atyp_angina'] == 1:  # atypical Angina
        angina[x] = 2
    elif heart_new.loc[x, 'ind_non_ang_pain'] == 1:  # non Angina
        angina[x] = 3
    else:
        angina[x] = 0  # asymptomatic
heart_new['Angina'] = angina
heart_new = heart_new.drop(['ind_typ_angina', 'ind_atyp_angina', 'ind_non_ang_pain'], axis=1)
# heart_new
```

```python
# ST-segment peak slop 1ope，-1down，0flat
slop = heart_new['ind_for_slope_up_exerc']
slop = slop.copy()
for x in heart_new.index:
    if heart_new.loc[x, 'ind_for_slope_down_exerc'] == 1:
        slop[x] = -1
heart_new = heart_new.drop('ind_for_slope_down_exerc', axis=1)
heart_new['ind_for_slope_up_exerc'] = slop
```

```python
# Thalassemia types 1reversible 2fixed 0normal
thal = heart_new['Thal_rev_defect']
thal = thal.copy()
for x in heart_new.index:
    if heart_new.loc[x, 'Thal_fixed_defect'] == 1:
        thal[x] = 2
heart_new['Thal_rev_defect'] = thal
heart_new = heart_new.drop('Thal_fixed_defect', axis=1)
```

```python
# reindex
heart_new = heart_new.rename(columns={'Angina': 'Angina_type', 'ind_for_slope_up_exerc': 'ind_for_slope_exerc',
                                      'Thal_rev_defect': 'Thal_defect'})
heart_new = heart_new.reindex(['Age', 'Sex', 'resting_BP', 'Serum_cholest', 'blood_sugar_exc120',
                               'Max_heart_rate', 'ind_exerc_angina', 'Angina_type', 'ST_dep_by_exerc',
                               'ind_for_slope_exerc', 'num_vessels_fluro',
                               'Thal_defect', 'Disease'], axis=1)
heart_new.to_csv('heartDieasePreprocess.csv', index = False)
```

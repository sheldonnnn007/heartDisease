```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
# process
heart_df = pd.read_csv('heartDieasePreprocess.csv')
features = heart_df.drop(columns = ['Disease'])
targets = heart_df['Disease']
features.head()
```

```python
# Convert discrete data from 0,1,2 to real string representation

# Sex
features.loc[features['Sex']==0,'Sex'] = 'female'
features.loc[features['Sex']==1,'Sex'] = 'male'

# blood_sugar_exc120
features.loc[features['blood_sugar_exc120'] == 1,'blood_sugar_exc120'] = 'true'
features.loc[features['blood_sugar_exc120'] == 0,'blood_sugar_exc120'] = 'false'

# ind_exerc_angina 
features.loc[features['ind_exerc_angina'] == 1,'ind_exerc_angina'] = 'true'
features.loc[features['ind_exerc_angina'] == 0,'ind_exerc_angina'] = 'false'

# Angina_type 
features.loc[features['Angina_type'] == 1,'Angina_type'] = 'typical'
features.loc[features['Angina_type'] == 2,'Angina_type'] = 'atypical'
features.loc[features['Angina_type'] == 3,'Angina_type'] = 'non-anginal'
features.loc[features['Angina_type'] == 0,'Angina_type'] = 'asymptomatic'

# ind_for_slope_exerc
features.loc[features['ind_for_slope_exerc'] == 1,'ind_for_slope_exerc'] = 'up'
features.loc[features['ind_for_slope_exerc'] == 0,'ind_for_slope_exerc'] = 'flat'
features.loc[features['ind_for_slope_exerc'] == -1,'ind_for_slope_exerc'] = 'down'

# num_vessels_fluro 
features['num_vessels_fluro'].astype("object")

# Thal_defect
features.loc[features['Thal_defect'] == 0,'Thal_defect'] = 'normal'
features.loc[features['Thal_defect'] == 2,'Thal_defect'] = 'fixed'
features.loc[features['Thal_defect'] == 1,'Thal_defect'] = 'reversable'


features.head()
```

```python
features.to_csv('features.csv', index = False)
targets.to_csv('targets.csv', index = False)
```
